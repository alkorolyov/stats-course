```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
lmd = 1e-6
rv = stats.expon(scale = 1 / lmd)
rv.rvs()

x = np.linspace(1, 10)
y = rv.pdf(x)
plt.plot(x, y);
```

```python
sigma = 1
x = np.linspace(-3, 3, num=100)

for i, sigma in enumerate([0.5, 1, 2]):
    y = np.exp(- (x / sigma) * (x / sigma) )
    plt.plot(x, y, label=f'p={sigma:.2f}')

plt.legend()
plt.show()
```
