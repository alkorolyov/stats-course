```python
import matplotlib.pyplot as plt
import scipy.stats as stats
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 10))

p = 0.5
N = 100000

rv = stats.bernoulli(p)
size = 1000

idx = np.linspace(1, N, N, dtype=int)
y = np.empty((size, N))

for i in range(size):
    arr = rv.rvs(N)
    y[i, :] = np.cumsum(arr) / idx


plt.plot(y.T, 'bx', markersize=2)
plt.xscale('log');


# plt.plot(y)
```
