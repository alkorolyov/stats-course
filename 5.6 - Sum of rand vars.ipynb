```python
import scipy.stats as stats
import numpy as np
import matplotlib.pyplot as plt

x = stats.uniform(0, 1)
```

```python
N = 10000

arr = []
for i in range(N):
    arr.append(x.rvs())

plt.hist(arr, bins=32);
```

```python
N = 100000

arr = np.zeros(N)

n = 512
for i in range(n):
    arr += np.random.binomial(1, 0.5, N) / n

plt.hist(arr, bins=n, histtype='step', alpha=0.5, label=f'sum {n}');

# arr = np.zeros(N)
# for i in range(128):
#     arr += np.random.binomial(1, 0.5, N) / 128
# plt.hist(arr, bins=100, alpha=0.5, label='sum 128');
plt.legend()
plt.show()
```

```python
""" Sum of two Gaussians """

mu_1 = -2
mu_2 = 2
sigma = 1

N = 100000

x = np.random.normal(mu_1, sigma**2, N)
y = np.random.normal(mu_2, sigma**2, N)

plt.hist(x, bins=100, alpha = 0.5, label='x')
plt.hist(y, bins=100, alpha = 0.5, label='y')
plt.hist(x + y, bins=100, alpha = 0.5, label='x+y')
plt.legend()
plt.show()
```
